```
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
import matplotlib.pyplot as plt
```

```
#We define energy production and electricity demand data as NumPy arrays
energy_production = np.array([40, 80, 50, 100, 60, 90, 70, 120, 30, 110, 140, 50, 160, 20, 130, 170, 90, 160, 70, 180, 110, 140, 100])
electricity_demand = np.array([35, 70, 45, 85, 55, 75, 65, 95, 25, 90, 110, 40, 120, 15, 100, 130, 80, 115, 60, 125, 85, 105, 80])
```

```
# Reshape the energy production data to use as input (X) in the model
x = energy_production.reshape(-1, 1)
y = electricity_demand
```

```
# We split the data into a training set and a test set (80% training, 20% testing)
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=42)
```

```
# We create a linear regression model
model = LinearRegression()
```

```
# We train the model with the training data
model.fit(x_train, y_train)
```

```
# We make predictions with test data
y_pred = model.predict(x_test)
```

```
# We show the model coefficients
print(f"Coefficient (Slope): {model.coef_[0]}")
print(f"Interception: {model.intercept_}")
```

```
# We visualize the results
plt.scatter(x_test, y_test, color='blue', label='Valores reales')
plt.plot(x_test, y_pred, color='red', label='Predicciones')
plt.xlabel('energy production (MW)')
plt.ylabel('Electricity demand (MW)')
plt.title('Linear regression - energy production vs Electricity demand')
plt.legend()
plt.show()
```

```
# We evaluate the model with the R^2 metric
r_squared = model.score(x_test, y_test)
print(f"R^2: {r_squared}")
```
